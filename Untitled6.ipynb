```python
import pandas as pd

data = {
    "Name": ["Rahul", "Anita", "Kiran", "Sneha", "Arjun"],
    "Maths": [85, 92, 70, 88, 60],
    "Physics": [78, 88, 65, 90, 55],
    "Chemistry": [90, 95, 72, 85, 58]
}

df = pd.DataFrame(data)
print(df)
```

```python
df["Average"] = df[["Maths", "Physics", "Chemistry"]].mean(axis=1)

def grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    else:
        return "D"

df["Grade"] = df["Average"].apply(grade)
print(df)

```

```python
top_student = df.loc[df["Average"].idxmax()]
print(top_student)
```

```python
import matplotlib.pyplot as plt

df.plot(x="Name", y="Average", kind="bar", legend=False)
plt.title("Student Average Marks")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.show()
```

```python
df.to_csv("output.csv", index=False)
```
